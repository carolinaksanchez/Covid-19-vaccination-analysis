```python
import numpy as np
house = ['1', '2','3', '4','5']
salePrice = [889000, 559000, 100000, 381900, 100000]
justValue = [851774, 453283, 210317, 563543, 212286]

arange = np.arange(len(house))
plt.xticks(arange, house)
plt.bar(arange-0.2, salePrice, width = 0.4, color= 'red', alpha = 0.3, label = "Sales price")
plt.bar(arange+0.2, justValue, width = 0.4, color= 'black', alpha = 0.5, label = "Just value")
plt.xlabel('House number')
plt.ylabel('Price')
plt.title('Houses Sold value vs Just value')
plt.legend()
```

The bar chart is showing the comparision between the first five houses last sold price and thier just value price.
