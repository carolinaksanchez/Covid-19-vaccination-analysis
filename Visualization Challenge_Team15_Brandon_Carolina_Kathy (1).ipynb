## Scatterplot

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline

path = 'C:\\Users\\kjhm\\Documents\\QMB3302\\Data Sets\\vaccination data.csv'
seg = pd.read_csv(path)
seg.head(43439)
```

```python
seg.dropna()
```

```python
sns.lmplot(x='total_vaccinations_per_hundred', y='people_fully_vaccinated_per_hundred', data=seg)
```

The scatterplot shown below compares total vaccinations to fully vaccinations and the strong positive correlation that the majority of people who get vaccinated in general are getting fully  vaccinated. With our immense amount of data points, this is one the graphs that we found to be visually representative of our data.

## Jointplot

```python
sns.jointplot(x='total_vaccinations_per_hundred', y='people_fully_vaccinated_per_hundred', data=seg, kind = 'scatter')
```

The below joint plot is an adaptation of our scatterplot but also shows the vast prevalence of data points in the earlier range of the x and y axis (total vaccination per hundred vs. people fully vaccinated per hundred). Not immediately apparent in our previous scatterplot, as the whole graph looks dense, most countries have quite a low vaccination rate per hundred, but with an increasing amount of vaccination in general, the higher the rate of which people become fully vaccinated. The statistical point this joint plot raises is the existence of a large amount of countries who are just not getting a high percentage of people vaccinated per population parameters.
